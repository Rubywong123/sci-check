```python
prompt_template = """Title: {}

Abstract: {}

Question: {} True, False, or Neither?

Answer: """
```

```python
import json
def list_to_dict(xs, keyname):
    return {x[keyname]: x for x in xs}

def load_jsonl(fname, max_lines=None):
    res = []
    for i, line in enumerate(open(fname)):
        if max_lines is not None and i == max_lines:
            return res
        else:
            res.append(json.loads(line))

    return res


claims = load_jsonl("data/claims.jsonl")
corpus = load_jsonl("data/corpus_candidates.jsonl")
```

```python
print(corpus[0])
print(claims[0])
```

```python
claim_dict = list_to_dict(claims, "id")
corpus_dict = list_to_dict(corpus, "doc_id")
```

```python
import random
for i in range(20):
    prompt = prompt_template
    claim = random.choice(claims)
    if i % 5 == 0:
        paper_id = int(list(claim['evidence'].keys())[0]) if len(claim['evidence'])> 0 else (random.choice(corpus))['doc_id']

    else:
        paper_id = (random.choice(corpus))['doc_id']
    if str(paper_id) in claim['evidence']:
        print(claim['evidence'][str(paper_id)]['label'])
    else:
        print("NEI")
    paper = corpus_dict[paper_id]
    
    abstract_counter = [x + 1 for x in range(len(paper["abstract"]))]
    abstract_sents = [
        f"[{counter}] {sentence}"
        for counter, sentence in zip(abstract_counter, paper["abstract"])
    ]
    abstract_sents = " ".join(abstract_sents)
    
    prompt = prompt.format(paper['title'], abstract_sents, claim['claim'])
    print(prompt)
```
